```python
import os

def clean_unlisted_images(base_folder, keep_list):
    """
    Elimina todas las imágenes en base_folder y sus subdirectorios
    excepto las que están en keep_list.
    
    :param base_folder: Carpeta raíz donde se encuentran las imágenes a revisar.
    :param keep_list: Lista de rutas absolutas de imágenes que se deben conservar.
    """
    # Convertir la lista de imágenes a un set para búsqueda rápida
    keep_set = set(keep_list)
    # Recorrer la carpeta y eliminar las imágenes no listadas
    counter = 0
    for root, _, files in os.walk(base_folder):
        for file in files:
            file_path = os.path.join(root, file)
            if file_path not in keep_set:
                os.remove(file_path)  # Eliminar el archivo
                #print(f"Eliminado: {file_path}")
                counter = counter + 1
    print(f'{counter} imagenes eliminadas')

# Definir la carpeta base y la lista de imágenes permitidas
base_folder = "/app/output/reveal/CFD_one_shot/CFD_one_shot_steguz/transformations"
# Ejecutar la limpieza
clean_unlisted_images(base_folder, path_images_to_keep)

#app\output\reveal\CFD_one_shot\CFD_one_shot_steguz\stego\AF-200/CFD-AF-200-228-N_recovered
#'/app/output/reveal/CFD_one_shot/CFD_one_shot_steguz/stego/AF-200/CFD-AF-200-228-N_recovered.png'
```

```python
import os

def count_images_in_folder(base_folder):
    """
    Cuenta el número total de imágenes en base_folder y sus subdirectorios.
    
    :param base_folder: Ruta de la carpeta raíz donde se encuentran las imágenes.
    :return: Número total de imágenes encontradas.
    """
    image_count = 0

    # Recorrer la carpeta y contar archivos con extensiones de imagen
    for root, _, files in os.walk(base_folder):
        image_count += len(files)  # Suma el número de archivos en la carpeta

    return image_count

# Definir la carpeta base
base_folder = "/app/output/reveal/CFD_one_shot/CFD_one_shot_steguz/transformations"

# Contar imágenes
total_images = count_images_in_folder(base_folder)
print(f"Total de imágenes en la carpeta: {total_images}")
```
